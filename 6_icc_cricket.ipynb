```python
ranking=womencricket[0::5]
```

```python
ranks=[]
for i in ranking:
    x=i.text
    ranks.append(x)
```

```python
matches=womencricket[2::5]
```

```python
matches[0].text
```

```python
matc=[]
for i in matches:
    x=i.text
    matc.append(x)
```

```python
name=womencricket[1::5]
```

```python
name[0].text.replace("\n","")
```

```python
naam=[]
for i in name:
    x=i.text.replace("\n","")
    naam.append(x)
```

```python
naam
```

```python
points=womencricket[3::5]
```

```python
poin=[]
for i in points:
    x=i.text.replace("\n","")
    poin.append(x)
```

```python
poin
```

```python
rate=womencricket[4::5]
```

```python
ratings=[]
for i in rate:
    x=i.text.replace("/n","").replace("\n","")
    ratings.append(x)
```

```python
ratings
```

```python
df_cricket=pd.DataFrame({"Ranking":ranks,"Team":naam,"Matches":matc,"Points":poin,"Ratings":ratings})
```
