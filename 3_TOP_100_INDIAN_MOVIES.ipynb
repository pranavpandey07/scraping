```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/india/top-rated-indian-movies/"
output=requests.get(url)
```

```python
output.status_code
```

```python
soup=BeautifulSoup(output.content,"html.parser")
```

```python
print(soup)
```

```python
movie_name=[]
rating=[]
year=[]
```

```python
movie_data=soup.findAll("td",class_="titleColumn")
```

```python
movie_data
```

```python
type(movie_data)
```

```python
movie_data=soup.findAll("a")
```

```python
movie_data
```

```python
atag=[]
for i in movie_data:
    atag.append(i)
```

```python
atag
```

```python
new_list=atag[63::2]
```

```python
new_list
```

```python
for i in new_list:
    x=i.text
    movie_name.append(x)
```

```python
movie_name=movie_name[0:250]
```

```python
len(movie_name)
```

```python
movie_rating=soup.find_all("td",class_="ratingColumn")
```

```python
movie_rating
```

```python
movie_rating[0]
```

```python
movie_rating[1]
```

```python
movie_rating[2]
```

```python
movie_rating[3]
```

```python
new_rat=movie_rating[0::2]
```

```python
for element in new_rat:
    x=element.text.replace("\n","")
    rating.append(x)
```

```python
len(rating)
```

```python
movie_year=soup.find_all('span',class_="secondaryInfo")
```

```python
movie_year[0].text.replace("(","").replace(")","")
```

```python
for i in movie_year:
    x=i.text.replace("(","").replace(")","")
    year.append(x)
```

```python
year
```

```python
len(year)
```

```python
df_movie=pd.DataFrame({"Name of Movie":movie_name,"Year of Release":year,"Ratings":rating})
```
