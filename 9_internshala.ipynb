```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
url="https://internshala.com/fresher-jobs/analytics,data%20science,machine%20learning-jobs"
output=requests.get(url)
```

```python
output
```

```python
soup=BeautifulSoup(output.content,"html.parser")
```

```python
print(soup)
```

```python
job_title=soup.find_all("div",class_="heading_4_5 profile")
```

```python
job_title
```

```python
extract=job_title[0].text.replace("\n","")
```

```python
extract
```

```python
job=[]
for i in job_title:
    x=i.text.replace("\n","")
    job.append(x)
    
```

```python
job
```

```python
comp=soup.find_all("div",class_="heading_6 company_name")
```

```python
comp[0].text.replace("\n","")
```

```python
company=[]
for i in comp:
    x=i.text.replace("\n","")
    company.append(x)
```

```python
company
```
