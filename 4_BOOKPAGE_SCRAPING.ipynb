```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
url="https://bookpage.com/search/?utf8=%E2%9C%93&query=romantic"
output=requests.get(url)
```

```python
output
```

```python
soup=BeautifulSoup(output.content,"html.parser")
```

```python
print(soup)
```

```python
book_name=soup.find_all("h4",class_="italic")
```

```python
name=[]
```

```python
for i in book_name:
    x=i.text.replace("\n","")
    name.append(x)
```

```python
name
```

```python
author=[]
author_name=soup.find_all("p",class_="sans bold")
```

```python
author_name[0].text.replace("\n","")
```

```python
for i in author_name:
    x=i.text.replace("\n","")
    author.append(x)
```

```python
author
```

```python
genre=[]
```

```python
genre_books=soup.find_all("p",class_="genre-links hidden-phone")
```

```python
genre_books
```

```python
for i in genre_books:
    x=i.text.replace("\n","").replace("YA","").replace("/","")
    genre.append(x)
```
