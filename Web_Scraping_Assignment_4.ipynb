# Calling the webdriver and putting in the url.

```python
driver=webdriver.Chrome("chromedriver.exe")
url="https://archive.ics.uci.edu/"
driver.get(url)
```

# Creating empty lists for various attributes of our dataframe and adding the data into the lists from website by using xpath method.

```python
all_dataset=driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a').click()
```

```python
name=[]
for i in range (2,590):
    nam=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[1]/table/tbody/tr/td[2]/p/b/a')
    for i in nam:
        name.append(i)
        
names=[]
for i in name:
    names.append(i.text)
```

```python
datatype=[]
for i in range (2,590):
    dat=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[2]/p')
    for i in dat:
        datatype.append(i)
        
datatyps=[]
for i in datatype:
    datatyps.append(i.text)
```

```python
task=[]
for i in range (2,590):
    ta=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[3]/p')
    for i in ta:
        task.append(i)
        
tasks=[]
for i in task:
    tasks.append(i.text)
```

```python
attr=[]
for i in range (2,590):
    at=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[4]/p')
    for i in at:
        attr.append(i)
        
attr_typ=[]
for i in attr:
    attr_typ.append(i.text)
```

```python
num_i=[]
for i in range (2,590):
    num=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[5]/p')
    for i in num:
        num_i.append(i)
        
number_i=[]
for i in num_i:
    number_i.append(i.text)
```

```python
num_a=[]
for i in range (2,590):
    num=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[6]/p')
    for i in num:
        num_a.append(i)
        
number_a=[]
for i in num_a:
    number_a.append(i.text)
```

```python
year=[]
for i in range (2,590):
    yea=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[7]/p')
    for i in yea:
        year.append(i)
        
years=[]
for i in year:
    years.append(i.text)
```

# Creating the final dataframe with the help of our lists.

```python
df_uci=pd.DataFrame({"Name":names,"Data Types":datatyps,"Default tasks":tasks,"Attribute Types":attr_typ,"Number of instances":number_i,"Number of attributes":number_a,"Year":years})
```
